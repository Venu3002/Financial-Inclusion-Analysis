```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
databank=pd.read_excel(r"C:\Users\GIM\Downloads\Dataset\Dataset\Data\CountryLevel_DatabankWide.xlsx")
microworld=pd.read_csv(r"C:\Users\GIM\Downloads\Dataset\Dataset\Data\micro_world_139countries.csv",encoding='latin-1')
print("MICROWORLD INFORMATION:")
print(microworld.info())
print(microworld.head(3))
print("DATABANK INFORMATION")
print(databank.info())
print(databank.head(3))
```

```python
mw_no=microworld.select_dtypes(include=['number']).columns
print("Microworld Numeric columns:",mw_no)

mw_str=microworld.select_dtypes(include=['object']).columns
print("Microworld String columns:", mw_str)

db_no=databank.select_dtypes(include=['number']).columns
print("DataBank Numeric columns:",db_no)

db_str=databank.select_dtypes(include=['object']).columns
print("DataBank String columns:", mw_str)
      
      
```

```python
microworld[mw_no]=microworld[mw_no].fillna(microworld[mw_no].mean())
databank[db_no]=databank[db_no].fillna(databank[db_no].mean())
```

```python
print("IN MICROWORLD")
print(microworld.isnull().sum())
print("\n IN DATABANK")
print(databank.isnull().sum())
```

```python
from scipy.stats import f_oneway
primary=microworld[microworld["educ"]==1]["account_fin"]
secondary=microworld[microworld["educ"]==2]["account_fin"]
tertiary=microworld[microworld["educ"]==3]["account_fin"]
anova_stat , anova_p = f_oneway(primary,secondary,tertiary)
print(f"ANOVA Statistic:{anova_stat:.2f},P-Value:{anova_p:.5f}")
if p_value<0.05:
    print("There is a significant impact in financial account ownership based on education level")
else:
    print("No Significant difference based on education level")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
X = microworld.drop(columns=['account_fin'])
Y = microworld['account_fin']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
log_model = LogisticRegression()
log_model.fit(X_train, Y_train)
Y_pred = log_model.predict(X_test)
accu=accuracy_score(Y_test,Y_pred)
print(f"Logistic Regression Accuracy : {accu:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Train vs test accuracy
tra_acc=log_model.score(X_train_scaled,Y_train)
tes_acc=log_model.score(X_test_scaled,Y_test)

print(f"Train Accuracy: {tra_acc:.4f}")
print(f"Test Accuracy: {tes_acc:.4f}")
```

```python
#accuracy is very low in this model
#use random forest method
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
rf.fit(X_train_scaled, Y_train)
Y_pred_rf = rf.predict(X_test_scaled)

accu_rf = accuracy_score(Y_test, Y_pred_rf)
print(f"Random Forest Accuracy: {accu_rf:.4f}")
```

```python
rftra_acc=rf.score(X_train_scaled,Y_train)
rftes_acc=rf.score(X_test_scaled,Y_test)

print(f"Train Accuracy: {rftra_acc:.4f}")
print(f"Test Accuracy: {rftes_acc:.4f}")
```

```python
from sklearn.metrics import confusion_matrix
Y_pred_final = rf.predict(X_test_scaled)
Conmat=confusion_matrix(Y_test, Y_pred_final)
plt.figure(figsize=(5,5))
sns.heatmap(Conmat, annot=True, fmt="d", cmap="Blues",xticklabels=["No Account","Has Account"],yticklabels=["No Account","Has Account"])
plt.xlabel("Predicted trend")
plt.ylabel("Actual trend")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import classification_report
print("Classificiation Report :\n",classification_report(Y_test,Y_pred_final))
```

```python
from sklearn.metrics import roc_curve, auc
Y_probs = rf.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, _ = roc_curve(Y_test, Y_probs)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}", color="blue")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray") # Random classifier baseline
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
